#Problem 1 - cutlets.csv

```python
import pandas as pd
import numpy as np
from scipy import stats
import scipy.stats as norm
```

```python
Q1 = pd.read_csv("Cutlets.csv")
```

```python
Q1.head()
```

2 sample 2 tailed test - independent samples

H0 - Mu1 = Mu2 - No significant diffference in cutlet size,

H1 - Mu1 != Mu2 - Significant Difference in cutlet size

```python
#2 sample 2 tailed test - independent samples
pvalue = stats.ttest_ind(Q1['Unit A'],Q1['Unit B'])
pvalue
```

Pvalue > 0.05 = Accept Null Hypothesis.

No significant difference in the cutlet sizes

#Question 2 - LabTAT.csv

```python
Q2 = pd.read_csv("LabTAT.csv")
Q2.head()
```

Anova Test

H0 - all 4 recorded TATs are same - Mu1 = Mu2 = Mu3 = M4

H1 - Atleast one sample is different

```python
p_value = stats.f_oneway(Q2["Laboratory 1"], Q2["Laboratory 2"], Q2["Laboratory 3"], Q2["Laboratory 4"])
p_value
```

p value < 0.05 - Reject Null hypothesis.

Atleast one sample mean is different
