```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
```

```python
alco= pd.read_csv("wine.csv")
```

```python
print(alco.describe())
alco.head()
```

```python
alco
```

```python
alco['Type'].value_counts()
```

```python
alco= alco.iloc[:,1:]
alco
```

```python
alco.shape
```

```python
alco.info()
```

```python
alco_arry=alco.values
alco_arry
```

```python
alco_norm=scale(alco_arry)
alco_norm
```

```python
pca = PCA()
pca_values = pca.fit_transform(alco_norm)
pca_values
```

```python
pca.components_
```

```python
var = pca.explained_variance_ratio_
var
```

```python
Var = np.cumsum(np.round(var,decimals= 4)*100)
Var
```

```python
plt.plot(Var,color="orange");
```

```python
final_df=pd.concat([alco['Alcohol'],pd.DataFrame(pca_values[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
import seaborn as sns
fig=plt.figure(figsize=(10,5))
sns.scatterplot(data=final_df);
```

```python
sns.scatterplot(data=final_df, x='PC1', y='PC2', hue='Alcohol');
```

# # Hierarchial Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
plt.figure(figsize=(10,8))
dendrogram=sch.dendrogram(sch.linkage(alco_norm,'complete'))
```

```python
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(alco_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
alco2=alco.copy()
alco2['clustersid']=hclusters.labels_
alco2
```

## kmeans clustering


```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range (1,6):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(alco_norm)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS');
```

```python
clusters3=KMeans(3,random_state=30).fit(alco_norm)
clusters3
```

```python
clusters3.labels_
```

```python
alco3=alco.copy()
alco3['clusters3id']=clusters3.labels_
alco3
```
