```python
a = 1-stats.norm.cdf(38,loc=df.MPG.mean(), scale=df.MPG.std())
round(a*100,3)
```

```python
b = stats.norm.cdf(40,loc=df.MPG.mean(), scale=df.MPG.std())
round(b*100,3)
```

```python
c = stats.norm.cdf(20,loc=df.MPG.mean(), scale=df.MPG.std())
d = stats.norm.cdf(50,loc=df.MPG.mean(), scale=df.MPG.std())
round((d-c)*100,3)
```
